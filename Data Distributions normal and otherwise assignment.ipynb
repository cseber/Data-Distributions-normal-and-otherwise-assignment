```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
normal = np.random.normal(0, 1, 100)
plt.hist(normal, color='c')
plt.axvline(np.mean(normal), color = 'b', linestyle='solid', linewidth=2)
plt.axvline(np.mean(normal) - np.std(normal), color = 'r', linestyle='dashed', linewidth=2)
plt.axvline(np.mean(normal) + np.std(normal), color = 'r', linestyle='dashed', linewidth=2)
plt.show()
```

For a normal distribution around half of the data is above and below the mean. Around 2/3 of the data falls between +/- 1 standard deviation.

```python
poisson = np.random.poisson(1,100)
plt.hist(poisson, color='c')
plt.axvline(np.mean(poisson), color = 'b', linestyle='solid', linewidth=2)
plt.axvline(np.mean(poisson) - np.std(poisson), color = 'r', linestyle='dashed', linewidth=2)
plt.axvline(np.mean(poisson) + np.std(poisson), color = 'r', linestyle='dashed', linewidth=2)
plt.show()
```

The mean and standard deviation do not prove useful for a poisson distribution.

```python
rand1 = np.random.normal(5, .5, 100)
rand2 = np.random.normal(10, 1, 100)
rand3 = rand1 + rand2
plt.hist(rand3, color='c')
plt.axvline(x=np.mean(rand3), color='blue', linewidth=2, linestyle='solid')
plt.axvline(x=np.mean(rand3) - np.std(rand3), color='red', linewidth=2, linestyle='dashed')
plt.axvline(x=np.mean(rand3) + np.std(rand3), color='red', linewidth=2, linestyle='dashed')
plt.show()
```

The mean and standard deviation do not seem to prove useful for this histogram.
